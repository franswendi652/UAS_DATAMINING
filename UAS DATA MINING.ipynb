```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Contoh data (pastikan x_train dan y_train terdefinisi dengan benar dalam konteks Anda)
x_train = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y_train = np.array([0, 0, 1, 1])

# Menggunakan contoh classifier sederhana untuk prediksi
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)

x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min()-1, stop = x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start = x_set[:, 1].min()-1, stop = x_set[:, 0].max() + 1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=ListedColormap(['red', 'green']))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c=ListedColormap(['red', 'green'])(i), label=j)
plt.title('Klasifikasi Data dengan Naive Bayes')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

# Contoh data
x = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([0, 0, 1, 1])

# Membagi data menjadi training dan testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

# Melatih model
classifier = GaussianNB()
classifier.fit(x_train, y_train)

# Menyiapkan data untuk plotting
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))

# Plotting decision boundary
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=ListedColormap(['red', 'green']))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

# Plotting points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c=ListedColormap(['red', 'green'])(i), label=j)

# Menambahkan detail plot
plt.title('Klasifikasi Data dengan Naive Bayes')
plt.xlabel('Umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [1, 2000],
    [2, 3000],
    [3, 4000],
    [4, 5000],
    [5, 6000],
    [6, 7000],
    [7, 8000],
    [8, 9000],
    [9, 10000],
    [10, 11000]
])
# Mengurutkan data berdasarkan Estimasi Gaji (kolom kedua)
sorted_data = data[data[:, 1].argsort()]

# Memilih 5 Estimasi Gaji tertinggi
top_5_data = sorted_data[-5:]

# Membuat diagram lingkaran untuk 5 Estimasi Gaji tertinggi
labels = [f'Umur {int(age)}' for age in top_5_data[:, 0]]
sizes = top_5_data[:, 1]
colors = ['red', 'blue', 'green', 'orange', 'purple']

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('5 Estimasi Gaji Tertinggi')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Contoh data
data = np.array([
    [1, 2000],
    [2, 3000],
    [3, 4000],
    [4, 5000],
    [5, 6000],
    [6, 7000],
    [7, 8000],
    [8, 9000],
    [9, 10000],
    [10, 11000]
])

# Mengurutkan data berdasarkan estimasi gaji (kolom kedua)
sorted_data = data[data[:, 1].argsort()]

# Memilih 5 estimasi gaji tertinggi
top_5_data = sorted_data[-5:]

# Warna untuk setiap batang
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Membuat grafik bar dengan warna
plt.bar(range(1, 6), top_5_data[:, 1], tick_label=top_5_data[:, 0], color=colors)
plt.xlabel('ID')
plt.ylabel('Estimasi Gaji')
plt.title('5 Estimasi Gaji Tertinggi')
plt.show()
```
